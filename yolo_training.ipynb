Inference 

```python
from ultralytics import YOLO
import os
import cv2

# -----------------------
# PATHS
# -----------------------
MODEL_PATH = "/content/drive/MyDrive/yolotrain/runs/train/weights/best.pt"
INPUT_DIR = "/content/drive/MyDrive/yolotrain/test_images"
OUTPUT_DIR = "/content/drive/MyDrive/yolotrain/output"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# -----------------------
# LOAD MODEL
# -----------------------
model = YOLO(MODEL_PATH)

# -----------------------
# RUN INFERENCE
# -----------------------
for img_name in os.listdir(INPUT_DIR):
    img_path = os.path.join(INPUT_DIR, img_name)

    results = model(img_path)

    # save result image with boxes
    for r in results:
        output_path = os.path.join(OUTPUT_DIR, img_name)
        r.save(filename=output_path)

print("✅ Inference Completed")
```
